```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the color palette
custom_colors = [
    '#FFE6CC', '#D5E8D4', '#F8CECC', '#BAC8D3', '#CCE5FF',
    '#DAE8FC', '#FFF2CC', '#B0E3E6', '#FAD7AC', '#D0CEE2',
    '#FFCCFF', '#FF9999', '#E6FFCC', '#CCFFFF', '#CCE5FF'
]

# Set a general style
sns.set(style="whitegrid")

# Function to plot bar charts
def plot_bar(data, column, title):
    plt.figure(figsize=(8, 5))
    order = data[column].value_counts().index
    sns.countplot(data=data, y=column, order=order, palette=custom_colors)
    plt.title(title)
    plt.xlabel("Count")
    plt.ylabel(column.capitalize())
    plt.tight_layout()
    plt.show()

# Barplots for Gender, Race, Ethnicity, and View
plot_bar(df, 'sex', 'Gender Distribution')
plot_bar(df, 'race', 'Race Distribution')
plot_bar(df, 'ethnicity', 'Ethnicity Distribution')
plot_bar(df, 'view', 'View Distribution')
```

```python
def make_bar_plot(data, column, title, xlabel, ylabel, palette="coolwarm", rotateLabels=False, removeDuplicatePatientIds=True):
    
    if removeDuplicatePatientIds==True:
        # Remove duplicate rows based on the 'id' column (keep the first occurrence)
        data = df.drop_duplicates(subset='patient-number', keep='first')
    
    # Create bar plot
    ax = sns.countplot(x=column, data=data, palette=palette)

    # Labels & title
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)

    # Add value labels on top of bars
    for p in ax.patches:
        ax.annotate(f"{p.get_height()}",  # Get bar height
                    (p.get_x() + p.get_width() / 2, p.get_height()),  # Position label
                    ha='center', va='bottom', fontsize=12, fontweight='bold')  # Formatting

    if rotateLabels == True:
        # Adjust x-tick labels by adding space
        plt.xticks(rotation=30, ha='right', va='top')  # Rotate and adjust label alignment
        
        
    plt.show()
```

```python
#make_bar_plot(data, column, title, xlabel, ylabel, palette="coolwarm"):
make_bar_plot(df, "sex", "Gender Distribution", "Gender", "Count")
```
